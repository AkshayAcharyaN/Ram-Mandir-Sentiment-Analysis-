```python
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.util import ngrams
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
def preprocess_text(text):
    tokens = word_tokenize(text.lower())  # Lower the text 

    tokens = [re.sub(r'[^a-zA-Z]', '', token) for token in tokens if token.isalpha()] # creating token 

    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]    # Removing the Stopword 

    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]   # Lemmatizing the token 

    return tokens

def generate_ngrams(tokens, n):
    return list(ngrams(tokens, n))
```
